### 멜론 사이트 접속

### 함수 만들기

```python
def melon_music_crawling(chrome_path):
    driver = webdriver.Chrome(chrome_path)
    url = 'https://www.melon.com/chart/index.htm'
    driver.get(url)
    driver.maximize_window()
    
    soup = bs(driver.page_source, 'lxml')
    
    # 데이터 수집
    song_info = soup.find_all('div', 'wrap_song_info')
    
    # 곡이름, 가수이름 분리
    song_and_singer = song_info[::2]
    
    # 앨범 이름 분리
    album = song_info[1::2]
    
    # 곡 이름 리스트 만들기
    song_name = []
    for ss in song_and_singer:
        temp = ss.find('div', 'ellipsis rank01')
        temp = temp.text
        song_name.append(temp.strip())
    
    # 가수 이름 리스트 만들기
    singer_name = []
    for ss in song_and_singer:
        temp = ss.find('span', 'checkEllipsis')
        temp = temp.text
        singer_name.append(temp.strip())
    
    # 앨범 리스트 만들기
    album_name =[]
    for temp in album:
        temp = temp.find('div', 'ellipsis rank03')
        temp = temp.text.strip()
        album_name.append(temp.strip())
    
    print(
        "수집된 곡 수: ", len(song_name), '\n',
        "수집된 가수이름 수: ", len(singer_name), '\n',
        "수집된 앨범 수: ", len(album_name), '\n')
    
    
    melon_music_df = pd.DataFrame({'순위': ranking_list,
                              '곡': song_name,
                              '앨범': album_name,
                              '가수': singer_name})

    print("Dataframe shape:", melon_music_df.shape)
    
    return melon_music_df
```

```python
chrome_path = 'C:/Users/User/Desktop/chromedriver/chromedriver_win32/chromedriver'

melon = melon_music_crawling(chrome_path)
```
