# 소리바다 데일리 인기차트 크롤링 함수 만들기

```python
def soribada_crawler(chromedriver_path):
    driver = webdriver.Chrome(chromedriver_path)
    url = 'https://www.soribada.com/music/chart'
    driver.get(url)

    # 검색 완료 후 창 최대화
    driver.maximize_window()

    # 1초 기다림
    time.sleep(1)
    
    soup = bs(driver.page_source,'lxml')
    frame = soup.find('ul', 'music-list')
    
    
    # 노래 제목 추출
    song_title_list = []
    for il in frame:
        # il: 전체 틀에서 1-100순위 곡이 들어있는 area 추출
        # area: 곡명, 아티스트, 앨범명 있는 area 추출
        area = il.find('div', 'list-area2')

        # 곡명만 있는 area 추출
        song_title_area = area.find('span', 'link-type1')

        # 곡명 추출
        song_title = song_title_area.find('span', 'song-title')

        # text로 변환
        song_title_text = song_title.text.strip()

        song_title_list.append(song_title_text)
        
    # 가수이름 추출
    artist_name_list = []
    for il in frame:
        # il: 전체 틀에서 1-100순위 곡이 들어있는 area 추출
        # area: 곡명, 아티스트, 앨범명 있는 area 추출
        area = il.find('div','list-area2')

        artist_name_span = area.find_all('span')
        artist_name = artist_name_span[6].text.strip()

        artist_name_list.append(artist_name)
        
    album_name_list = []
    
    # 앨범명 추출
    for il in frame:
        # il: 전체 틀에서 1-100순위 곡이 들어있는 area 추출
        # area: 곡명, 아티스트, 앨범명 있는 area 추출
        area = il.find('div', 'list-area2')

        album_name_area = area.find('span', 'link-type2 album-title')
        album_name_text = album_name_area.find('a')['title']

        album_name_list.append(album_name_text.strip())
    
    # 데이터프레임 만들기
    rank = list(range(1, 101))
    
    soribada_daily_df = pd.DataFrame({"순위": rank,
                                  "노래제목": song_title_list,
                                  "가수명": artist_name_list,
                                  "앨범명": album_name_list})

    soribada_daily_df.head()
    
    return soribada_daily_df
```
